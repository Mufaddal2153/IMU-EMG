```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('dataset.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
x = data["X"]
y = data["Y"]
z = data["Z"]
```

```python
z.head()
```

```python
temp_data = data.drop("Mixed",axis=1)
```

```python
X = temp_data.drop("ClassLabel",axis=1)
y = data['ClassLabel']
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

k_range = 3
scores = {}
scores_list = []
knn = KNeighborsClassifier(n_neighbors=k_range)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
```

```python
y_pred = knn.predict(X_test)
metrics.accuracy_score(y_pred,y_test)*100
```

```python
sns.set(font_scale=1.5)
conf = confusion_matrix(y_test,y_pred)
plt.title("Confusion matrix of KNN plotted on heatmap")
sns.heatmap(conf,annot=True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
mse = mean_squared_error(y_test, y_pred)
rmse = sqrt(mse)
rmse
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

from sklearn import tree
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
acc=accuracy_score(y_test,y_pred)*100
print(f"The accuracy of the Decision Tree Model is {acc}")
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100)
```

```python
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
acc=accuracy_score(y_test,y_pred)*100
print(f"The accuracy of the Random Forest Model is {acc}")
```

```python
sns.set(font_scale=1.5)
conf = confusion_matrix(y_test,y_pred)
plt.title("Confusion matrix of Random Forest plotted on heatmap")
sns.heatmap(conf,annot=True)
plt.show()
```
